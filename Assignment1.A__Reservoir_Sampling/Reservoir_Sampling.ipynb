```python
import random
import os
import pandas as pd

def sample(file, K): # Parameters: the file and the list size
    reservoir_Samples = [] # Use a  simple list to store the elements

    with open(file, 'r') as file: # Open a file and read every line one by one
        for i in range(K): # First use a list(<==>array) to store the first K items
            line = file.readline()  # Read a line

            if not line: # if our file has < K lines terminate
                break

            reservoir_Samples.append(line)

        # In case we have remaining items in the stream(more than K)
        i = K # Keep the lines read
        while 1:
            line = file.readline()
    
            if not line:
                break 

            i = i+1 # keep the counter updated
            r = random.randint(1, i) # Pick a random index from 1 to i
            if r <= K: # Only replace if r us valid (from 1 to K)
                reservoir_Samples[r-1] = line # Replace item at index r-1 with new item

    file.close()
    return reservoir_Samples  # Return the list of the randomly selected items of the stream
```

```python
# main

file = 'input.txt'  # The file that contains the stream of items
K = 20 # Number of items to store in the list(<==> sizez of the list)
sampled_items = sample(file, K)

# Print the output
print("Selected items produced from the Reservoir Sampling Algorithm:")
for item in sampled_items:
    print(item)
```
